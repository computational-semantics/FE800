```python
import glob
import errno
import spacy
import heapq
import textacy
import re
import pandas as pd
import numpy as np
nlp = spacy.load('en_core_web_lg')
```

```python
path = 'D:/FE/FE800/Data/finance-sample/*.txt' #note C:
files = glob.glob(path)

df=[]
for name in files:
    try:
        with open(name, encoding="utf8") as f:
            li1=[]
            for line in f:
                li1.append(line)
            df.append(" ".join(li1))
                
    except IOError as exc: #Not sure what error this is
        if exc.errno != errno.EISDIR:
            raise
```

```python
df1 = []
df2 = []
i=0

for d in range(len(df)):
    words = textacy.preprocess_text(df[d], fix_unicode=False, lowercase=True, no_urls=True, no_emails=True, no_phone_numbers=True, no_numbers=True, no_currency_symbols=True, no_punct=True, no_contractions=True, no_accents=True)
    wordss = textacy.preprocess.normalize_whitespace(words)
    df1.append(wordss)
    doc = nlp(wordss)
    j = 0
    k = 0
    for j in range(len(doc)):
        if k != 500:
            if doc[j].vector_norm != 0:
                df2.append(doc[j].vector)
                k += 1
            else:
                continue
        else:
            break
    i=i+1
    print(i)
```

```python
df3 = []
temp = []
for l in range(len(df2)):
    temp2 = []
    for h in range(len(df2[l])):
        temp.append(df2[l][h])
    if (l+1) %500 == 0:
        df3.append(temp)
        temp = []
```

```python
df4 = pd.DataFrame(df3)
```

```python
df4.to_csv('finance-sample.csv', index = False, header = False)
```
